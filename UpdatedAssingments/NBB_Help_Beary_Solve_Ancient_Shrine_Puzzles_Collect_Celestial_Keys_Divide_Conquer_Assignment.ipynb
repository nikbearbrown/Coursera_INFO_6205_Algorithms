### Solution Template:
```python
from typing import List

def solve_puzzles(puzzles: List[int]) -> int:
    puzzles.sort()
    total_time = 0
    current_time = 0
    for puzzle in puzzles:
        current_time += puzzle
        total_time += current_time
    return total_time
```
